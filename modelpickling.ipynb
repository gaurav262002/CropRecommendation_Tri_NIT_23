```python
import pandas as pd
```

```python
df = pd.read_csv('agricult-data.csv')
```

```python
df.head()
```

```python
df.commodity.unique()
```

```python
crop_season = {
'apple': 'K',
'banana': 'K',
'coconut': 'K',
'cotton': 'K',
'grapes': 'R',
'jute': 'Z',
'maize': 'K',
'mango': 'R',
'orange': 'K',
'papaya': 'K',
'pomegranate': 'Z',
'rice': 'K',
}
print(crop_season)
```

```python
df['season'] = df['commodity']
df["season"].replace(crop_season, inplace=True)
```

```python
df.head()
```

```python
df.state.unique()
```

```python
df.season.unique()
```

```python
state_dict = {}
count = 0
for ele in list(df.state.unique()):
    state_dict[ele] = count
    count+=1
state_dict
```

```python
season_dict = {}
count = 0
for ele in list(df.season.unique()):
    season_dict[ele] = count
    count+=1
season_dict
```

```python
crop_dict = {}
count = 0
for ele in list(df.commodity.unique()):
    crop_dict[ele] = count
    count+=1
crop_dict
```

```python
result_dict = dict([(value, key) for key, value in crop_dict.items()])
result_dict
```

```python
df["state"].replace(state_dict, inplace=True)
df["season"].replace(season_dict, inplace=True)
df["commodity"].replace(crop_dict, inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
X = df[['state','season','modal_price']]
Y = df['commodity']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 4625)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth=8)
dt.fit(X_train, y_train)
```

```python
y_train_pred = dt.predict(X_train)
y_test_pred = dt.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print(accuracy_score(y_test, y_test_pred))
```

```python
import pickle
pickle.dump(dt, open('model.pkl', 'wb'))
```
